## Boston Housing Polynomial Regression

```python
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
import sys
import matplotlib.pyplot as plt 
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
bd = pd.read_csv('Boston.csv')
bd = bd.loc[:, ~bd.columns.str.contains('^Unnamed')]
X = bd['dis'].values
y = bd['nox'].values
```

## Degree - 2

```python
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X.reshape(-1,1))
lr = LinearRegression()
lr.fit(X_poly,y)

X_fit = np.arange(X.min(),X.max(),1)[:,np.newaxis]

y_predict = lr.predict(poly_reg.fit_transform(X_fit.reshape(-1,1)))

plt.scatter(X,y)
plt.plot(X_fit,y_predict)
```

## Cubic

```python
poly_reg_2 = PolynomialFeatures(degree=3)
X_poly = poly_reg_2.fit_transform(X.reshape(-1,1))
lr = LinearRegression()
lr.fit(X_poly,y)

X_fit = np.arange(X.min(),X.max(),1)[:,np.newaxis]

y_predict = lr.predict(poly_reg_2.fit_transform(X_fit.reshape(-1,1)))
y_predict_1 = lr.predict(poly_reg_2.fit_transform(np.array([10]).reshape(-1,1)))
print(y_predict_1)
plt.scatter(X,y)
plt.plot(X_fit,y_predict)
```
