```python
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
import sys
import matplotlib.pyplot as plt 
```

```python
bd = pd.read_csv('Boston.csv',index_col=0)
```

```python
bd.head()
```

```python
bd.describe()
```

```python
sns.pairplot(bd,height=1.5)
plt.show()
```

```python
col_study = ['crim','zn','rm','indus','medv','chas']
sns.pairplot(bd[col_study],height=2.0)
```

```python
sns.color_palette("mako", as_cmap=True)
sns.heatmap(bd[col_study].corr(),annot=True)
plt.show()
```

```python
bd.head()
```

```python
X = bd['rm'].values.reshape(-1,1)
```

```python
y = bd['medv'].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.figure(figsize=(12,10))
sns.regplot(x=X,y=y)
plt.xlabel('avg rooms')
plt.ylabel('median')
plt.show()
```

```python
model.predict(np.array([9]).reshape(-1,1))
```
